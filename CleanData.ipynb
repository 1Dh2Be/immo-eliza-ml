```python
# load neccessary modules
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 

# load data
df = pd.read_csv("data/input.csv")

# Take columns of interest only (can add more later on)
columns_of_interest = [
    'price', 
    'property_type',
    'province',
    'zip_code',
    'subproperty_type',
    'total_area_sqm',
    'nbr_bedrooms',
    'fl_terrace',
    'fl_garden',
    'state_building',
    'primary_energy_consumption_sqm',
]

new_df = df[columns_of_interest]

# fill missing columns here
mean_sqm_per_category = df.groupby('property_type')['total_area_sqm'].mean()

new_df['total_area_sqm'] = df.apply(
    lambda row: mean_sqm_per_category[row['property_type']] if pd.isna(row['total_area_sqm']) else row['total_area_sqm'],
    axis=1
)

median_energy_consumption_sqm = new_df.groupby(["property_type", "zip_code"])["primary_energy_consumption_sqm"].transform("median")

new_df['primary_energy_consumption_sqm'].fillna(median_energy_consumption_sqm, inplace=True)

new_df.info()

# new_df.dropna(subset=["province"], inplace=True)
```

```python
# Qualitative -> Quantitative
new_df["property_type"] = new_df['property_type'].replace({'HOUSE': 1, 'APARTMENT': 0}) 

new_df = pd.get_dummies(new_df, columns=["subproperty_type"])

new_df = pd.get_dummies(new_df, columns=["province"])

new_df = pd.get_dummies(new_df, columns=["state_building"])
# Converts true/false to 1/0
new_df = new_df * 1


# Calculate the correlation coefficient
correlation = new_df['total_area_sqm'].corr(new_df['price'])

# Print the correlation coefficient
print(f"Correlation between 'feature' and 'price': {correlation}")
```

```python
# Set display options to show all columns
pd.set_option('display.max_columns', None)

# Drop col
new_df = new_df.drop("province_MISSING", axis=1)

# Displays the first 5 rows of the data set
new_df.head(25)
```
